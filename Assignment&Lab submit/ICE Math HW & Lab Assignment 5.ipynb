```python
#1)

import numpy as np
import matplotlib.pyplot as plt
from cmath import pi

function1 = [10, 0, 0, 0, 0]
y1 = np.poly1d(function1) # 4*x**4 

print(y1)
```

```python
#1-a) y'

y1prime = np.polyder(y1)

print(y1prime)
```

```python
#1-b)
xvals = list(np.arange(-1,1,0.1))

yvals = []
for x in xvals:
    yvals.append(y1(x))  

yprimevals = []
for x in xvals:
    yprimevals.append(y1prime(x))  

plt.plot(xvals, yvals, 'r-')
plt.plot(xvals, yprimevals, 'b-')

plt.title('red : y, blue :y\'')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#2

from sympy import *
from cmath import pi


x = Symbol('x')
y2 = (x**2)*(exp(x))
print("y =",y2)
y2integral = integrate(y2)-2
print("z =",y2integral)


f1 = lambdify(x, y2, 'numpy')
z = np.array(np.linspace(0, 1, 100))
plt.plot(z, f1(z), 'g')

f2 = lambdify(x, y2integral, 'numpy')
z = np.array(np.linspace(0, 1, 100))
plt.plot(z, f2(z), 'r')

plt.title('red : z, green : y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#3

x = np.array(np.linspace(0, 2*pi, 101))
y = np.cos(x)
yprime = np.diff(y)/np.diff(x)
x = x[0:len(x)-1]
plt.title('red: y\'=-sinx, blue: y=cosx, green : -sinx')

plt.plot(x, yprime, 'or', x, np.cos(x), 'ob', x , -np.sin(x), 'xg')
```

```python
delx = 0.1

t = np.arange(0, 2.1, delx)
a = [0, 9.05, 16.37, 22.22, 26.81, 30.33, 32.93, 34.76, 35.95, 36.59, 36.79, 36.62, 36.14, 35.43, 34.52, 33.47, 32.30, 31.06, 29.75, 28.42, 27.07]
v = delx*np.cumsum(a)
d = delx*np.cumsum(v)
plt.title('red: velocity, blue: displacement')
plt.plot(t, v, '-r', t, d, '-b')
```
