```python
#1) Let c1=4+5j and c2=1-6j

from numpy import *
import cmath
import matplotlib.pyplot as plt

c1 = complex(4,5)
c2 = complex(1,-6)

#a) a = c1 - c2

a = c1-c2
print("a = c1 - c2 = ", a)

#b) b = c1*c2

b = c1*c2
print("b = c1*c2 = ", b)

#c) c = c1/c2 

c = c1/c2 
print("c = c1/c2 = ", c)


#d) The magnitude of c1

c1_mag = abs(c1)
print("The magnitude of c1 = ", c1_mag)

#e) The phase of c2

c2_pha = rad2deg(cmath.phase(c2))
print("The phase of c2 = ", c2_pha)
```

```python
#3) Let w = 10*pi, t = [0:0.01:5].  Also let p1 = 3exp(i*(wt+pi/3)), p2 = 5*exp(i*wt+pi/4)). 

w = 10*pi
t = arange(1, 5, 0.01)
p1 = 3*exp(1j*(w*t+pi/3))
p2 = 5*exp(1j*w*t+pi/4)
p3 = p1 + p2

#a) Plot p1 using 

X1 = [x.real for x in p1]
Y1 = [x.imag for x in p1]
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(X1, Y1, t, 'red')

```
