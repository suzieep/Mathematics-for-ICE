```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt
max_t=2*pi
samples=201
deltat=max_t/samples
x=10

A=10
delta=0.005
a_1=5
t=np.arange(-a_1,a_1,delta)

def XF(f,x,t,deltat):
    xf=sum(x*np.exp(-1j*f*t))*deltat
    return xf

Fx=[]
frange = np.arange(-2*pi,2*pi,1/2000,dtype=float)
for f in frange:
    xf=XF(f,x,t,delta)
    Fx.append(xf)
    
plt.plot(frange,np.absolute(Fx))
```
