```python
#1-a)cos(5t)

import numpy as np
from math import pi
import matplotlib.pyplot as plt

coe = 5
f = coe/(2*pi)
print("frequency = ", f) #frequency = 5/2pi

t = 1/f
xvals = np.linspace(-t, t, num=1000)
ycos = np.cos(coe*xvals) 

plt.plot(xvals, ycos, '-')
plt.title('cos(5t)')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([-t, t, -1, 1])
plt.grid()
plt.show()
```

```python
#1-b)cos(2pi(100t))


coe = 2*pi*100
f = coe/(2*pi)
print("frequency = ", f) #frequency = 2*pi*100/2pi = 100
t = 1/f
xvals = np.linspace(-t, t, num=1000)
ycos = np.cos(coe*xvals) 

plt.plot(xvals, ycos, '-')
plt.title('cos(2pi(100t))')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([-t, t, -1, 1])
plt.grid()
plt.show()
```

```python
#2-a)Plot mt

xvals = np.arange(0, 0.1, 0.0001)

mtcoe = 2*pi*10
mt = np.cos(mtcoe*xvals)
f = mtcoe/(2*pi)
print("frequency = ", f) #frequency = 2*pi*10/2pi = 10
t = 1/f



plt.plot(xvals[1:3000], mt[1:3000], '-')
plt.title('mt')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([0, 0.1, -1, 1])
plt.grid()
plt.show()
```

```python
#2-b)Plot ct


xvals = np.arange(0, 0.1, 0.0001)

ctcoe = 2*pi*300
ct = np.cos(ctcoe*xvals) 
f = ctcoe/(2*pi)
print("frequency = ", f) #frequency = 2*pi*300/2pi = 300
t = 1/f




plt.plot(xvals[1:3000], ct[1:3000], '-')
plt.title('ct')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([0, 0.1, -1, 1])
plt.grid()
plt.show()
```

```python
#2-c) Plot xc = mt*ct for xc

xvals = np.arange(0, 0.1, 0.0001)
ct = np.cos(ctcoe*xvals) 
mt = np.cos(mtcoe*xvals)
xc = mt*ct
#frequency = 300-10,300+10


plt.plot(xvals[1:3000], xc[1:3000], '-')
plt.title('xc')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([0, 0.1, -1, 1])
plt.grid()
plt.show()
```

```python
#2-d) Plot the spectrum of signal xc using the following commands:

xvals = np.arange(0, 1, 0.001)
ct = np.cos(ctcoe*xvals) 
mt = np.cos(mtcoe*xvals)
xc = mt*ct
#frequency = 300-10,300+10

spec = np.abs(np.fft.fft(xc));
N = len(spec)/2;
plt.plot(spec[1:500]/N)


plt.title('xc')
plt.xlabel('Input')
plt.ylabel('Function values')

plt.grid()
plt.show()
```

```python
#2-e) Plot xi=xc*ct for xi[1:3000].

xvals = np.arange(0, 0.1, 0.0001)
ct = np.cos(ctcoe*xvals) 
mt = np.cos(mtcoe*xvals)
xc = mt*ct
xi = xc*ct
#frequency = 300-10,300+10


plt.plot(xvals[1:3000], xi[1:3000], '-')
plt.title('xi')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.axis([0, 0.1, -1, 1])
plt.grid()
plt.show()
```

```python
#2-f)Plot the spectrum of signal xi using the following commands:

xvals = np.arange(0, 1, 0.0001)
ct = np.cos(ctcoe*xvals) 
mt = np.cos(mtcoe*xvals)
xc = mt*ct
xi = xc*ct

spec2 = np.abs(np.fft.fft(xi));
N = len(spec2)/2;
plt.plot(spec2[1:1000]/N)


plt.title('xi')
plt.xlabel('Input')
plt.ylabel('Function values')

plt.grid()
plt.show()
```
